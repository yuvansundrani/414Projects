```python
def perform_kmeans_clustering(data, k_value):
    features = data[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
    kmeans = KMeans(n_clusters=k_value, random_state=42)
    data['Cluster'] = kmeans.fit_predict(features)
    return data
```

```python
def analyze_clusters(data, k_value):
    print(data['Cluster'].value_counts())
    for cluster in range(k_value):
        print(f"\nCluster {cluster + 1}:")
        print(data[data['Cluster'] == cluster][['Name', 'Type 1', 'Type 2']])
```

```python
def analyze_clusters(data, k_value):
    print(data['Cluster'].value_counts())

    plt.figure(figsize=(10, 6))
    sns.countplot(x='Cluster', data=data)
    plt.title('Distribution of Data Points in Each Cluster')
    plt.xlabel('Cluster')
    plt.ylabel('Count')
    plt.show()

    for cluster in range(k_value):
        cluster_data = data[data['Cluster'] == cluster][['Name', 'Type 1', 'Type 2']]
        
        print(f"\nCluster {cluster + 1}:")
        display(cluster_data)
```

```python
file_path = "Pokemon.csv"
pokemon_data = load_data(file_path)

k_value = 4

pokemon_data = perform_kmeans_clustering(pokemon_data, k_value)

analyze_clusters(pokemon_data, k_value)
```
